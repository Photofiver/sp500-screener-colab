```python
import yfinance as yf
import pandas as pd
import numpy as np
import requests
from io import StringIO
from tqdm.notebook import tqdm
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)
pd.set_option('display.max_colwidth', 50)

print("✅ Importy zakończone pomyślnie")
```

```python
# ============================================================
# ============== 🔧 PARAMETRY ANALIZY (EDYTUJ TUTAJ!) ==============
# ============================================================

print("🔧 Ładowanie parametrów analizy...")

# ==================== FILTRY I PROGI ====================
MIN_MARKET_CAP_B      = 8.0      # Minimalna kapitalizacja (mld USD)
MAX_PE_RATIO          = 28.0     # Maksymalne P/E
MIN_ROE_PCT           = 10.0     # Minimalne ROE (%)
MIN_PROFIT_MARGIN_PCT = 7.0      # Minimalna marża zysku (%)
MIN_REVENUE_GROWTH_PCT = 4.0     # Minimalny wzrost przychodów (%)
MAX_DEBT_TO_EQUITY    = 2.0      # Maksymalny dług / kapitał własny

# ==================== RSI ====================
RSI_PERIOD      = 14
RSI_OVERBOUGHT  = 70             # Powyżej tego = penalizacja

# ==================== WAGI PUNKTOWE (suma ≈ 100) ====================
W_MARKET_CAP     = 12   # Wielkość = stabilność
W_PE             = 18   # Wartość (niskie P/E)
W_ROE            = 18   # Jakość biznesu (rentowność)
W_PROFIT_MARGIN  = 10   # Efektywność
W_REVENUE_GROWTH = 14   # Wzrost
W_DEBT           = 10   # Bezpieczeństwo bilansu
W_VOLUME_GROWTH  = 5    # Momentum wolumenu
W_RSI            = 13   # Sygnał techniczny

print(f"Parametry załadowane pomyślnie.")
print(f"   Min Market Cap: ${MIN_MARKET_CAP_B}B | Max PE: {MAX_PE_RATIO} | Min ROE: {MIN_ROE_PCT}%")
```

```python
def get_sp500_tickers():
    """Pobiera aktualną listę tickerów S&P 500 z Wikipedii (z User-Agent)"""

    url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8"
    }
    try:
        resp = requests.get(url, headers=headers, timeout=15)
        resp.raise_for_status()
        tables = pd.read_html(StringIO(resp.text))
        wiki_df = tables[0]
        tickers = wiki_df['Symbol'].dropna().astype(str).tolist()
        # Standaryzacja dla yfinance
        tickers = [t.replace('.', '-') for t in tickers]
        print(f"✅ Pobrano {len(tickers
)} tickerów S&P 500 z Wikipedii.")
        return tickers
    except Exception as e:
        print(f"❌ Błąd pobierania listy: {e}")
        return []

def calculate_rsi(prices: pd.Series, period: int = 14) -> float:
    """Oblicza klasyczny RSI"""

    if len(prices) < period + 5:
        return 50.0
    delta = prices.diff()
    gain = delta.where(delta > 0, 0.0).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0.0)).rolling(window=period).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    val = rsi.iloc[-1]
    return float(val) if not pd.isna(val) else 50.0

def calculate_score(data):
    """Prosty scoring 0-100 na podstawie wielu czynników"""

    score = 0.0

    # 1. Market Cap
    mc = data.get('Market_Cap_B', 0)
    if mc >= 200: score += W_MARKET_CAP
    elif mc >= 50: score += W_MARKET_CAP * 0.75
    elif mc >= MIN_MARKET_CAP_B: score += W_MARKET_CAP * 0.45

    # 2. P/E (im niższe tym lepiej)
    pe = data.get('PE_Ratio', 999)
    if pe <= 12: score += W_PE
    elif pe <= 18: score += W_PE * 0.75
    elif pe <= MAX_PE_RATIO: score += W_PE * 0.4

    # 3. ROE
    roe = data.get('ROE_Pct', 0)
    if roe >= 25: score += W_ROE
    elif roe >= 16: score += W_ROE * 0.75
    elif roe >= MIN_ROE_PCT: score += W_ROE * 0.4

    # 4. Profit Margin
    pm = data.get('Profit_Margin_Pct', 0)
    if pm >= 18: score += W_PROFIT_MARGIN
    elif pm >= 10: score += W_PROFIT_MARGIN * 0.7
    elif pm >= MIN_PROFIT_MARGIN_PCT: score += W_PROFIT_MARGIN * 0.35

    # 5. Revenue Growth
    rg = data.get('Revenue_Growth_Pct', 0)
    if rg >= 12: score += W_REVENUE_GROWTH
    elif rg >= 7: score += W_REVENUE_GROWTH * 0.7
    elif rg >= MIN_REVENUE_GROWTH_PCT: score += W_REVENUE_GROWTH * 0.35

    # 6. Debt/Equity
    de = data.get('Debt_to_Equity', 999)
    if de <= 0.4: score += W_DEBT
    elif de <= 1.0: score += W_DEBT * 0.7
    elif de <= MAX_DEBT_TO_EQUITY: score += W_DEBT * 0.35

    # 7. Volume Growth (momentum)
    vg = data.get('Volume_Growth_Pct', 0)
    if vg >= 25: score += W_VOLUME_GROWTH
    elif vg >= 10: score += W_VOLUME_GROWTH * 0.6

    # 8. RSI
    rsi = data.get('RSI', 50)
    if rsi <= 35: score += W_RSI * 0.9
    elif rsi <= 55: score += W_RSI * 0.7
    elif rsi <= RSI_OVERBOUGHT: score += W_RSI * 0.3

    return min(round(score, 1), 100.0)

def get_grade(score):
    if score >= 82: return 'A+'
    if score >= 72: return 'A'
    if score >= 62: return 'B+'
    if score >= 52: return 'B'
    if score >= 42: return 'C+'
    if score >= 32: return 'C'
    if score >= 22: return 'D'
    return 'F'

print("✅ Funkcje pomocnicze zdefiniowane")
```

```python
print("
🚀 Rozpoczynam pełną analizę S&P 500...
")

tickers = get_sp500_tickers()

if not tickers:
    print("❌ Nie udało się pobrać listy tickerów. Kończę wykonanie.")
else:
    results = []
    error_count = 0

    for ticker in tqdm(tickers, desc="📈 Analizuję spółki S&P 500", unit="spółka"):
        try:
            stock = yf.Ticker(ticker)
            info = stock.info

            # Sprawdzenie czy dane fundamentalne istnieją
            if not info or info.get('marketCap', 0) < 1_000_000_000:
                error_count += 1
                continue

            # Dane historyczne (3 miesiące)
            hist = stock.history(period="3mo", auto_adjust=True, progress=False)
            if len(hist) < RSI_PERIOD + 10:
                continue

            # === Ekstrakcja danych ===
            market_cap_b = round(info.get('marketCap', 0) / 1_000_000_000, 2)
            pe_ratio = info.get('trailingPE') or info.get('forwardPE') or 999.0
            roe_pct = round((info.get('returnOnEquity') or 0) * 100, 2)
            profit_margin_pct = round((info.get('profitMargins') or 0) * 100, 2)
            revenue_growth_pct = round((info.get('revenueGrowth') or 0) * 100, 2)

            total_debt = info.get('totalDebt', 0) or 0
            equity = info.get('totalStockholderEquity', 1) or 1
            debt_equity = round(total_debt / equity, 2) if equity > 0 else 999.0

            # Volume growth
            vol = hist['Volume']
            vol_recent = vol.tail(15).mean()
            vol_prev = vol.iloc[-35:-15].mean() if len(vol) > 35 else vol_recent
            vol_growth_pct = round(((vol_recent - vol_prev) / vol_prev * 100), 2) if vol_prev > 100 else 0.0

            # RSI
            rsi_val = round(calculate_rsi(hist['Close'], RSI_PERIOD), 2)

            long_name = (info.get('longName') or info.get('shortName') or ticker)[:55]
            sector = info.get('sector', 'Unknown')

            record = {
                'Ticker': ticker,
                'Nazwa': long_name,
                'Sektor': sector,
                'Market_Cap_B': market_cap_b,
                'PE_Ratio': round(pe_ratio, 2) if pe_ratio < 500 else None,
                'ROE_Pct': roe_pct,
                'Profit_Margin_Pct': profit_margin_pct,
                'Revenue_Growth_Pct': revenue_growth_pct,
                'Debt_to_Equity': debt_equity if debt_equity < 500 else None,
                'Volume_Growth_Pct': vol_growth_pct,
                'RSI': rsi_val
            }

            record['Score'] = calculate_score(record)
            record['Grade'] = get_grade(record['Score'])
            results.append(record)

        except Exception as e:
            error_count += 1
            if error_count % 40 == 0:
                print(f"   Błędy: {error_count} (ostatni ticker: {ticker})")
            continue

    print(f"
✅ Analiza zakończona!")
    print(f"   Przeanalizowano spółek: {len(results)}")
    print(f"   Błędy / pominięte: {error_count}")

    if results:
        df = pd.DataFrame(results)
        df = df.sort_values('Score', ascending=False).reset_index(drop=True)
        df.insert(0, 'Rank', range(1, len(df) + 1))

        # TOP 20
        print("
" + "=" * 95)
        print("🏆 TOP 20 SPÓŁEK S&P 500 — NAJLEPSZY SCORING")
        print("=" * 95)

        display_cols = ['Rank', 'Ticker', 'Nazwa', 'Sektor', 'Market_Cap_B', 'PE_Ratio', 'ROE_Pct', 'Score', 'Grade']
        display(df[display_cols].head(20))

        # Statystyki
        print(f"
📊 Statystyki:")
        print(f"   Średni score: {df['Score'].mean():.1f}   |   Mediana: {df['Score'].median():.1f}")
        print(f"   Najlepsze oceny: {dict(df['Grade'].value_counts().head(5))}")

        # Zapis do Excela
        output_file = "SP500_Results.xlsx"
        df.to_excel(output_file, index=False, sheet_name='S&P500 Analysis', engine='openpyxl')
        print(f"
💾 Zapisano {len(df)} wierszy do pliku: {output_file}")

        # Automatyczne pobieranie w Google Colab
        try:
            from google.colab import files
            files.download(output_file)
            print("📥 Plik Excel został przygotowany do pobrania!")
        except ImportError:
            print("ℹ️  (Aby pobrać plik — uruchom notebook w Google Colab)")

    else:
        print("❌ Brak wyników do wyświetlenia.")
```
